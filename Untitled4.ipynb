```python
import pandas as pd
import re
from collections import Counter
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
file_path = 'train.txt'  # or csv file if needed
df = pd.read_csv(file_path, sep=';', header=None, names=['Text', 'Emotion'])
print(df.head())
```

```python
def basic_tokenize(s):
    s = s.lower()
    s = re.sub(r'[^a-z0-9\s]+', '', s)
    return s.strip().split()
```

```python
df['Tokens'] = df['Text'].apply(basic_tokenize)
print(df.head())
```

```python
counter = Counter()
for tokens in df['Tokens']:
    counter.update(tokens)
itos = ['<pad>', '<unk>'] + [w for w, _ in counter.items()]
stoi = {w: i for i, w in enumerate(itos)}
vocab_size = len(itos)
pad_idx = stoi['<pad>']
unk_idx = stoi['<unk>']
print(itos[:10], vocab_size)
```

```python
def numericalize(tokens):
    return [stoi.get(tok, stoi['<unk>']) for tok in tokens]
df['Numericalized'] = df['Tokens'].apply(numericalize)
print(df.head())
```

```python
df['label'] = [i for i, label in enumerate(df['Emotion'].unique()) for _ in range(len(df[df['Emotion'] == label]))]
print(df['label'].head())
```

```python
class TextDataset(Dataset):
    def __init__(self, df):
        self.items = list(zip(df['Numericalized'], df['label']))
        self.label_map = {label: idx for idx, label in enumerate(df['Emotion'].unique())}

    def __getitem__(self, idx):
        tokens, label = self.items[idx]
        return tokens, label

    def __len__(self):
        return len(self.items)
```

```python
def pad_collate(batch, pad_idx=0):
    max_len = max(len(ids) for ids, _ in batch)
    padded = [
        ids + [pad_idx] * (max_len - len(ids)) for ids, _ in batch
    ]
    labels = [label for _, label in batch]
    return torch.tensor(padded), torch.tensor(labels)
```

```python
dataset = TextDataset(df)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True, collate_fn=lambda x: pad_collate(x, pad_idx=pad_idx))

for batch in dataloader:
    tokens, labels = batch
    print("Tokens:", tokens)
    print("Labels:", labels)
    break
```

## Load and preprocess test data

```python
test_file_path = 'test.txt'
df_test = pd.read_csv(test_file_path, sep=';', header=None, names=['Text', 'Emotion'])

df_test['Tokens'] = df_test['Text'].apply(basic_tokenize)
df_test['Numericalized'] = df_test['Tokens'].apply(numericalize)
df_test['label'] = df_test['Emotion'].apply(lambda x: stoi.get(x, -1))

test_dataset = TextDataset(df_test)
test_dataloader = DataLoader(test_dataset, batch_size=4, shuffle=False, collate_fn=lambda x: pad_collate(x, pad_idx=pad_idx))

print(df_test.head())
```

```python
label_map = {label: idx for idx, label in enumerate(df['Emotion'].unique())}
df_test['label'] = df_test['Emotion'].map(label_map)

test_dataset = TextDataset(df_test)
test_dataloader = DataLoader(test_dataset, batch_size=4, shuffle=False, collate_fn=lambda x: pad_collate(x, pad_idx=pad_idx))

print(df_test.head())
```

## Define cnn model

### Subtask:
Define a simple CNN model architecture for text classification using PyTorch.


**Reasoning**:
Define the CNN model architecture for text classification using PyTorch.



```python
import torch.nn as nn
import torch.nn.functional as F

class TextCNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, num_filters, kernel_sizes, num_classes, dropout=0.5):
        super().__init__()

        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=pad_idx)

        self.convs = nn.ModuleList([
            nn.Conv1d(in_channels=embedding_dim, out_channels=num_filters, kernel_size=k)
            for k in kernel_sizes
        ])

        self.fc = nn.Linear(len(kernel_sizes) * num_filters, num_classes)

        self.dropout = nn.Dropout(dropout)

    def forward(self, text):
        # text = [batch size, seq len]

        embedded = self.embedding(text)
        # embedded = [batch size, seq len, embedding dim]

        embedded = embedded.permute(0, 2, 1)
        # embedded = [batch size, embedding dim, seq len]

        conved = [F.relu(conv(embedded)) for conv in self.convs]
        # conved_n = [batch size, num filters, seq len - kernel_size + 1]

        pooled = [F.max_pool1d(conv, conv.shape[2]).squeeze(2) for conv in conved]
        # pooled_n = [batch size, num filters]

        cat = self.dropout(torch.cat(pooled, dim=1))
        # cat = [batch size, num filters * len(kernel_sizes)]

        return self.fc(cat)

embedding_dim = 100
num_filters = 100
kernel_sizes = [3, 4, 5]
num_classes = len(df['Emotion'].unique())

model = TextCNN(vocab_size, embedding_dim, num_filters, kernel_sizes, num_classes)

print(model)
```

## Prepare for training

### Subtask:
Set up the loss function, optimizer, and move the model to the appropriate device (CPU or GPU).


**Reasoning**:
Set up the loss function, optimizer, and move the model to the appropriate device.



```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

print(f'Using device: {device}')
```

## Train the model

### Subtask:
Implement the training loop to train the CNN model for 5 epochs.


**Reasoning**:
Implement the training loop for 5 epochs.



```python
num_epochs = 5

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for tokens, labels in dataloader:
        tokens, labels = tokens.to(device), labels.to(device)

        optimizer.zero_grad()

        predictions = model(tokens)

        loss = criterion(predictions, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f'Epoch {epoch+1}, Loss: {total_loss/len(dataloader)}')
```

## Evaluate the model

### Subtask:
Evaluate the trained model's accuracy on the test dataset.


**Reasoning**:
Evaluate the trained model's accuracy on the test dataset by iterating through the test dataloader, making predictions, and comparing them to the true labels.



```python
model.eval()
correct_predictions = 0
total_samples = 0

with torch.no_grad():
    for tokens, labels in test_dataloader:
        tokens, labels = tokens.to(device), labels.to(device)

        predictions = model(tokens)
        predicted_classes = torch.argmax(predictions, dim=1)

        correct_predictions += (predicted_classes == labels).sum().item()
        total_samples += labels.size(0)

accuracy = correct_predictions / total_samples
print(f'Test Accuracy: {accuracy:.4f}')
```
